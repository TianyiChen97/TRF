```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import vonmises
import scipy.stats
import random
import ruptures as rpt
from sklearn.metrics import confusion_matrix
import warnings
warnings.filterwarnings('ignore')
```

```python
# About the simulate data:
# Four time point: wakeup, first-meal, last-meal, sleep
# Monday to Friday are around 6am, 8am, 6pm, 9pm
# Saturday and Sunday are around 7am, 9am, 7pm, 10pm
# Noise control in 0.25h
# penalty = 1
# 6 months, 24 weeks, 24*7=168 days
```

```python
wakeup = []
for i in range(24):
    for j in range(5):
        numberweekday = round(random.uniform(5.75,6.25), 2)
        wakeup.append(numberweekday)
        j += 1
    for k in range(2):
        numberweekend = round(random.uniform(6.75,7.25), 2)
        wakeup.append(numberweekend)
        k += 1
    i += 1
    

firstMeal = []
for i in range(24):
    for j in range(5):
        numberweekday = round(random.uniform(7.75,8.25), 2)
        firstMeal.append(numberweekday)
        j += 1
    for k in range(2):
        numberweekend = round(random.uniform(8.75,9.25), 2)
        firstMeal.append(numberweekend)
        k += 1
    i += 1
    

lastMeal = []
for i in range(24):
    for j in range(5):
        numberweekday = round(random.uniform(17.75,18.25), 2)
        lastMeal.append(numberweekday)
        j += 1
    for k in range(2):
        numberweekend = round(random.uniform(18.75,19.25), 2)
        lastMeal.append(numberweekend)
        k += 1
    i += 1
    

sleep = []
for i in range(24):
    for j in range(5):
        numberweekday = round(random.uniform(20.75,21.25), 2)
        sleep.append(numberweekday)
        j += 1
    for k in range(2):
        numberweekend = round(random.uniform(21.75,22.25), 2)
        sleep.append(numberweekend)
        k += 1
    i += 1
```

```python
df = pd.read_csv("start.csv")
df['wakeup'] = wakeup
df['firstMeal'] = firstMeal
df['lastMeal'] = lastMeal
df['sleep'] = sleep
df
```
